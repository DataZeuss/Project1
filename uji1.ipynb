```python
import re
import pandas as pd
table = pd.read_csv("dataset.csv")
bahanbakar = table['Mileage'] 
bahanbakarsep = []

for bb in bahanbakar:
    inputan = re.split("( )",bb)
    bahanbakarsep.append(inputan)
len(bahanbakarsep)
```

```python
del bahanbakarsep[0]
bahanbakarsep = np.array(bahanbakarsep)
nilai, satuan = [], []
for i in range (len(bahanbakarsep)):
    nilai.append(bahanbakarsep[i,0])
    satuan.append(bahanbakarsep[i,2])
print(nilai)
print(satuan)
```

```python
bbkmpl, bbkmkg = [], []
for i in range (len(satuan)):
    if satuan[i] == 'kmpl':
        bbkmpl.append(nilai[i])
    else:
        bbkmkg.append(nilai[i])
print(bbkmpl)
print(bbkmkg)
```
